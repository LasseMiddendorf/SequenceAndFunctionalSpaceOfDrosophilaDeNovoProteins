```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
TMs = pd.read_csv("denovos_AFDB50_Tm.csv", index_col=0)
rmsd = pd.read_csv("avg_RMSD.csv", index_col=0)

df = pd.merge(rmsd, TMs, how="inner", on="id")
```

```python
rmsd
```

```python
fig, ax = plt.subplots()

ax.hist(df["avg_RMSD"], density=True, cumulative=True, histtype='step', bins=100, color="#43978D")

ax.set_yticks([0, 0.25, 0.5, 0.75, 1])

ax.set_xlabel("Average RMSD of Trajectory [Ã…]")
ax.set_ylabel(r"Fraction of $\it{de}$ $\it{novo}$ proteins")
ax.set_xlim(0, 45)
ax.set_ylim(0, 1.05)

plt.savefig("RMSD_hist.pdf", dpi=600)
```

```python
fig, ax = plt.subplots()

ax.hist(df["avg_RMSD"], color="#43978D", bins=35, edgecolor="black")

ax.set_xlabel("Average RMSD of Trajectory [Ã…]")
ax.set_ylabel("Count")

plt.savefig("RMSD_hist_count.pdf", dpi=600)
```
