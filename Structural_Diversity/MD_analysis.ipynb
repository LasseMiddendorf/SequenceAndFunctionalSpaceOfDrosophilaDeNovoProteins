## Plot example RMSD trajectories

```python
droso_1087 = pd.read_csv("Data/droso_1087_rmsd.csv", index_col=False)
droso_1446 = pd.read_csv("Data/droso_1446_rmsd.csv", index_col=False)
droso_4840 = pd.read_csv("Data/droso_4840_rmsd.csv", index_col=False)

mpl.rcParams['axes.labelsize'] = 13
mpl.rcParams['xtick.labelsize'] = 13
mpl.rcParams["axes.titlesize"] = 15
mpl.rcParams["legend.fontsize"] = 13

def plot_data(df, ax):

    colors = ["#807f4b", "#dbbb5d", "#424967"]


    for i in range(1, 4):
        ax.plot(df["time"], df[f"replicate_{i}"], label=f"Replicate {i}", color=colors[i-1])
            
    ax.set_xlabel("Time [ns]")

    ax.set_ylim(0, 17)

    ax.set_xlim(0, 100)

    return ax
```

```python
#Create 1x3 subplots
fig, ax = plt.subplots(1, 3, figsize=(20, 5), sharey=True)

#Plot data

plot_data(droso_1087, ax[0])
plot_data(droso_1446, ax[1])
plot_data(droso_4840, ax[2])

#Add titles
ax[0].set_title("Droso_1087")
ax[1].set_title("Droso_1446")
ax[2].set_title("Droso_4840")

#Add y label
ax[0].set_ylabel("RMSD [Å]")

#Add legend
ax[2].legend(loc="center right", bbox_to_anchor=(1.45, 0.5))
```
